Data cleaning

```python
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
df.head()
```

Diseases Count

```python
disease_count=df['Condition'].value_counts()
print("\nDisease Count")
print(disease_count)
```

Age Group Count

```python
bins = [0,18,35,60,100]
labels=['0-18','19-35','36-60','60+']
df['Age_Group']=pd.cut(df['Age'],bins=bins,labels=labels)
age_count = df['Age_Group'].value_counts()
print("\nAge Group Count")
print(age_count)
```

Line Chat

```python

import pandas as pd
import matplotlib.pyplot as plt
df['Admission_Date'] = pd.to_datetime(df['Admission_Date'])
monthly = df.groupby(df['Admission_Date'].dt.month).size()
monthly.plot(kind='line',marker='o',color = "green",linestyle = "--")
plt.title('Monthly Admissions from  Data')
plt.xlabel('Month')
plt.ylabel('Number of Admissions')
plt.show()
```
