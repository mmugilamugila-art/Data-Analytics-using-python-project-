```python
Mark1 = int(input("Enter your first mark: "))
Mark2 = int(input("Enter your second mark: "))
Mark3 = int(input("Enter your third mark: "))
Mark4 = int(input("Enter your fourth mark: "))
Total = Mark1 + Mark2 + Mark3 + Mark4
print("Your total is: ", Total)
Average = Total / 4
print("Your average is: ", Average)
if Average >= 50: print("Pass")
else:
  print("Fail")
```

Grades

```python
Mark1 = int(input("Enter your first mark: "))
Mark2 = int(input("Enter your second mark: "))
Mark3 = int(input("Enter your third mark: "))
Mark4 = int(input("Enter your fourth mark: "))
Total = Mark1 + Mark2 + Mark3 + Mark4
print("Your total is: ", Total)
Average = Total / 4
print("Your average is: ", Average)
if Average >= 90:
   print("Grade A")
elif Average >= 80:
   print("Grade B")
elif Average >= 70:
   print("Grade C")
elif Average >= 60:
   print("Grade D")
else:
  print("Fail")
```
