```python
import numpy as np
sales = [87,65,69.99,88]

total_sales = sum(sales)
average_sales = total_sales/len(sales)
highest_sales = max(sales)
lowest_sales = min(sales)

print("Total Marks:",total_sales)
print("Average Marks:",average_sales)
print("Highest Marks:",highest_sales)
print("Lowest Marks:",lowest_sales)
```
