Clean a Messy Sales Dataset Day-6

```python
import pandas as pd
data = {
    "Employee_ID":[101,102,103,104,105],
    "Nmae":["Ananthu","Balaji","Chiththu","Devi","None"],
    "Age":[25,30,35,40,45],
    "Salary":[50000,60000,70000,80000,90000],
}
df = pd.DataFrame(data)
print(df)

print(df.isnull())

df = (df.drop_duplicates())
print(df)

df["Age"] = df["Age"].fillna(df["Age"].mean())
print(df)

df = df.dropna(subset=["Salary"])
print(df)

df = df.rename(columns={"Employee_ID":"Emp_ID"})
print(df)
```

groupby

```python
region_sales = df.groupby("Region")["Sales"].sum()

print(region_sales)
```

```python
product_sales = df.groupby("Product")["Sales"].sum()

print(product_sales)
```

Aggregation

```python
result = df.groupby("Region")["Sales"].agg(
    ["sum","mean","max","min"]
)

print(result)
```

Region-wise sales Report day 8

```python
import pandas as pd

data = {
    "Region":["North","South","East","West","North","South","East","West","North","South"],
    "Product":["Laptop","Mobile","Laptop","Tablet","Mobile","Tablet","Laptop","Mobile","Tablet","Laptop"],
    "Sales":[50000,25000,52000,18000,27000,19000,51000,26000,20000,55000],
    "Quantity":[5,10,4,6,9,7,5,11,8,5]
}

df = pd.DataFrame(data)

east_sales = df[df["Region"] == "East"]
print (east_sales)

sorted_df = df.sort_values("Sales")
print(sorted_df)

region_sales = df.groupby("Region")["Sales"].sum()
print(region_sales)

result = df.groupby("Region")["Sales"].agg(
    ["sum","mean","max","min","count"]
)
print(result)
```

Pie Chart

dashboard

```python
import matplotlib.pyplot as plt
months = ["jen","Feb","Mar","Apr","May","Jun"]
sales = [10000,12000,15000,13000,18000,20000]
plt.figure(figsize = (18,5))

#line chart
plt.subplot(1,3,1)

plt.plot(months,sales,marker ="o")
plt.title("Sales Trend")
plt.xlabel("Months")
plt.ylabel("sales")

#Bar Chart
plt.subplot(1,3,2)

plt.bar(months,sales,color ="orange")
plt.title("Sales Comperison")
plt.xlabel("Months")
plt.ylabel("sales")

#pie Chart
plt.subplot(1,3,3)
expenses = [6000,2000,4000]
categories = ["food","Travel","Shopping"]
plt.pie(expenses,
        labels = categories,
        autopct = "%1.1f%%", colors = ["green","Pink","Yellow"])
plt.title("Expenses Distribution")
plt.tight_layout()
plt.show()
```
