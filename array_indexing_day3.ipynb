array indexing

```python
import numpy as np
marks = np.array([47,76,43,75,60])
print(marks[3])
```

```python
import numpy as np
arr = np.array([1,2,3,4,5,6])
new_arr = arr.reshape(2,3)
print(new_arr)
```

```python
import numpy as np
marks = np.array([50,60,70,80.90])
average = np.mean(marks)
print(average)
```

```python
import numpy as np
marks = np.array([50,65,76,89.94])
average = np.median(marks)
print(average)
```
